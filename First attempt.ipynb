```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
#Data from vizier without treatment
raw_data = pd.read_csv(r"C:\Users\clauw\Documents\Programming\Final-Project\NIR_data_vizier.tsv", delimiter='\t',names=["Wave", "nFlux", "rFlux"])

#Convert the wavelength in nm cause its in cm^-1 
raw_data['Wave'] = (10**(7))/raw_data['Wave']

raw_data
```

```python
#Plot of raw data
fig1, axis1 = plt.subplots(2,1, figsize = (9, 7),)
ax1 = axis1[0]
ax2 = axis1[1]

fig1.suptitle(r'Vizier Data',fontsize=14)

ax1.plot(raw_data['Wave'],raw_data['nFlux'],color='orange')
ax1.set_xlabel('Wavelength (nm)',fontsize=14)
ax1.set_ylabel('Flux normalized',fontsize=14)
ax1.grid()
ax1.tick_params(axis='both',labelsize=13)

ax2.plot(raw_data['Wave'],raw_data['rFlux'],color='orange')
ax2.set_xlabel('Wavelength (nm)',fontsize=14)
ax2.set_ylabel('Flux raw',fontsize=14)
ax2.grid()
ax2.tick_params(axis='both',labelsize=13)

fig1.tight_layout()
```

```python
#We take out the absorption and only use the normalized flux
raw_data['nFluxInv'] = 1.0-raw_data['nFlux']

#Plot again
fig2, ax = plt.subplots(1,1, figsize=(12,7))
ax.plot(raw_data['Wave'],raw_data['nFluxInv'], color='orange')
ax.set_title('Vizier Data normalized and inverted')
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Flux')
ax.grid()
```
